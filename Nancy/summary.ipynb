```python
import pandas as pd

# Path to your raw CSV
RAW_PATH = (r"C:\Users\vande\becode\immo-eliza-cats-analysis\Nancy\sample_data_copy\properties.csv")

# Load CSV into a pandas DataFrame
df_raw = pd.read_csv(RAW_PATH)

# Quick look at first 5 rows
df_raw.head()
```

```python
# Automatically selects numeric columns
numeric_summary = df_raw.describe().T  # Transpose for readability

# Rename the 50% column to "median"
numeric_summary = numeric_summary.rename(columns={"50%": "median"})

# Show the summary
numeric_summary
```

```python
# Skewness
numeric_summary["skewness"] = df_raw[numeric_summary.index].skew()

# Kurtosis
numeric_summary["kurtosis"] = df_raw[numeric_summary.index].kurtosis()
numeric_summary["kurtosis"]
numeric_summary["skewness"]
```
