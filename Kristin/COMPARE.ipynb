```python
import pandas as pd

file1 = '../data/raw/immoweb_data.csv'
file2 = '../Kristin/cleaned_properties.csv'

df1 = pd.read_csv(file1, dtype=str)  # dtype=str to avoid numeric issues
df2 = pd.read_csv(file2, dtype=str)
```

```python
print("Shape of file1:", df1.shape)
print("Shape of file2:", df2.shape)
```

```python
if df1.equals(df2):
    print("Files are exactly the same.")
else:
    print("Files differ.")
```

```python
# Merge to find differences
merged = df1.merge(df2, how='outer', indicator=True)
diff_rows = merged[merged['_merge'] != 'both']
print(f"Number of differing rows: {len(diff_rows)}")
diff_rows.head(10)
```

```python
# Only consider rows where latitude and longitude are numeric
valid_coords = df1['latitude'].notna() & df1['longitude'].notna()

# Check for outliers within valid coordinates
lat_outliers = valid_coords & ~df1['latitude'].between(49.5, 52.5)
lon_outliers = valid_coords & ~df1['longitude'].between(2.5, 6.5)

# Count outliers
print("Rows with latitude outliers:", lat_outliers.sum())
print("Rows with longitude outliers:", lon_outliers.sum())

# View problematic rows
outliers = df1.loc[lat_outliers | lon_outliers, ['id', 'latitude', 'longitude', 'locality', 'region']]
outliers
```
