```python
import pandas as pd

# Load both files
raw = pd.read_csv(r'../data/raw/immoweb_data.csv')
cleaned = pd.read_csv(r'../Kristin/cleaned_properties.csv')  # adjust path

# Check unique values in swimming_pool
print("Raw unique values:", raw['fl_swimming_pool'].unique())
print("Cleaned unique values:", cleaned['fl_swimming_pool'].unique())

# Compare counts of each value
print("\nRaw counts:")
print(raw['fl_swimming_pool'].value_counts())
print("\nCleaned counts:")
print(cleaned['fl_swimming_pool'].value_counts())

# Optionally, check rows that were 1 in raw
missing_in_cleaned = raw[raw['fl_swimming_pool'] == 1].merge(
    cleaned, how='left', on='id', suffixes=('_raw', '_cleaned')
)
print("\nCheck swimming_pool values for rows that were 1 in raw:")
print(missing_in_cleaned[['id', 'fl_swimming_pool_raw', 'fl_swimming_pool_cleaned']])
```
